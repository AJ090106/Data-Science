```python
import sqlite3
```

```python
connection = sqlite3.connect("example.db")
connection
```

```python
cursor = connection.cursor()
```

```python
#create a table

cursor.execute('''
    Create table if not exists employees(
               id integer primary key,
               name text not null,
               age integer,
               department text
               )
''')

#commit the changes

connection.commit()
```

```python
#insert data in to the table

cursor.execute('''
    insert into employees(name,age,department)
               values('anuj',20,'software engineer')
''')


cursor.execute('''
    insert into employees(name,age,department)
               values('rudy',21,'data analyst')
''')


cursor.execute('''
    insert into employees(name,age,department)
               values('Smit',20,'Data Scientist')
''')


connection.commit()
```

```python
#query data from table

cursor.execute('Select * from employees')
rows = cursor.fetchall()

#printing the query
for row in rows:
    print(row)
```

```python
#update the data

cursor.execute('''
update employees
set age = 21
where name = 'anuj'
''')

connection.commit()
```

```python
#delete the data from the table

cursor.execute('''
delete from employees
where name = 'rudy'
''')
```
