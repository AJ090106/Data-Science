```python
#pie chart

labels = ['A','B','C','D']
sizes = [ 30,20,40,10]
colors = ['yellowgreen', 'lightcoral', 'lightskyblue', 'gold']
explode = (0.1,0,0,0)#move out the first slice

plt.pie(sizes,explode = explode,labels=labels,colors= colors,autopct="%1.1f%%")
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Generate dataset
n = 120

data = {
    "age": np.random.randint(18, 60, n),
    "salary": np.random.randint(25000, 120000, n),
    "experience": np.random.randint(0, 20, n),
    "department": np.random.choice(["Sales", "HR", "Tech", "Finance"], n),
    "performance_score": np.random.normal(70, 10, n).round(1),
    "work_hours": np.random.randint(4, 12, n)
}

df = pd.DataFrame(data)

# Save dataset as CSV
df.to_csv("employee_data.csv", index=False)

df.head()
```

```python
import seaborn as sns
```

```python
#basic plotting with seaborn
tips = sns.load_dataset("tips")
tips
```

```python
#create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y='tip',data=tips)
plt.title("total bill vs tip")
plt.show()
```

```python
#lineplot

sns.lineplot(x = 'size', y= 'total_bill', data = tips)
plt.title("Lineplot for size vs Total Bill")
plt.show()
```

```python
#categorical plots
sns.barplot(x = 'day',y= 'total_bill',data= tips)
plt.title('Bar plot for BIlls on the day')
plt.show()
```

```python
#boxplot

sns.boxplot(x= 'day', y='total_bill',data=tips)
plt.show()
```

```python
#Violin plot
sns.violinplot(x = 'day', y= 'total_bill',data=tips)
plt.show()
```

```python
sns.histplot(tips['total_bill'], bins = 10,kde = True)
plt.show()
```

```python
sns.kdeplot(tips['total_bill'],fill =True)
plt.show()
```

```python
sns.pairplot(tips)
```

```python
#heatmap

corr = tips[['total_bill','tip','size']].corr()
corr
sns.heatmap(corr,annot = True)
```
